```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot histograms for numeric columns
numeric_cols = ['item_price', 'quantity', 'transaction_amount']
df[numeric_cols].hist(bins=20, figsize=(10, 8))
plt.tight_layout()
plt.show()

# Plot box plots for numeric columns to identify outliers
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
# Sales over time
# Convert 'date' and 'time_of_sale' to datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['time_of_sale'] = pd.to_datetime(df['time_of_sale'], errors='coerce')

df['date'] = pd.to_datetime(df['date'])  # Ensure it's in datetime format
df.groupby(df['date'].dt.date)['transaction_amount'].sum().plot(figsize=(10, 6))
plt.title('Total Transaction Amount Over Time')
plt.xlabel('Date')
plt.ylabel('Total Transaction Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top-selling items by quantity
top_items_quantity = df.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(10)
top_items_quantity.plot(kind='bar', figsize=(10, 6))
plt.title('Top-Selling Items by Quantity')
plt.xlabel('Item Name')
plt.ylabel('Total Quantity Sold')
plt.show()

# Top items by transaction amount
top_items_amount = df.groupby('item_name')['transaction_amount'].sum().sort_values(ascending=False).head(10)
top_items_amount.plot(kind='bar', figsize=(10, 6))
plt.title('Top Items by Transaction Amount')
plt.xlabel('Item Name')
plt.ylabel('Total Transaction Amount')
plt.show()
```

```python


# Sales by item_type
plt.figure(figsize=(10, 6))
sns.boxplot(x='item_type', y='transaction_amount', data=df)
plt.title('Sales by Item Type')
plt.xticks(rotation=45)
plt.show()
```
